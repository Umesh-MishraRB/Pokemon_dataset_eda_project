# **How to fetch the dataset from the kaggle to the google notebook??**
We will use the kagglehub to get the dataset from the kaggle using the paths, and after that we will join the path with the csv.

```python
import kagglehub
import os
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
path=kagglehub.dataset_download("abcsds/pokemon")
print("The path of the dataset from the kaggle is: ", path)
data_path=os.path.join(path,"Pokemon.csv")
print(data_path)
df=pd.read_csv(data_path)
print(df.head(10))
```

# **How many pokemons exist with an attack value greater than 150?**

```python
print(df.loc[df['Attack']>150].shape)
row,_=df.loc[df['Attack']>150].shape
print("The number of pokemons with an attack value greater than 150 is: ",row)
```

# **Advanced selection**

Now let's use boolean operators to create more advanced expressions

## **How many fire-flying pokemons are there?**

```python
fire_flying_df=((df['Type 1']=="Fire") & (df["Type 2"]=="Flying")).sum()
print(fire_flying_df)
print(df.loc[(df["Type 1"]=="Fire")& (df["Type 2"]=="Flying")])
print((df.loc[(df["Type 1"]=="Fire")& (df["Type 2"]=="Flying")])["Name"])
```

# **How many poison pokemons are there across both the types?**

```python
print((df["Type 1"]==df["Type 2"]).sum())
print((df["Type 1"]=="Poison").sum())
# print(df["Type 1"].value_counts())
# print(df["Type 2"].value_counts())
print((df["Type 2"]=="Poison").sum())
print(((df["Type 1"]=="Poison") | (df["Type 2"]=="Poison")).sum())
```
